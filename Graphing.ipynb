```python
import math

value_list = []
# damage = (int) (damage * (100f / tempValue));
for i in range(0, 1000):
    current_value = 100 / (100 + i)
    value_list.append(current_value)
```

```python
from matplotlib import pyplot as plt

plt.plot(value_list)
plt.ylabel("Final Percentage of Damage Done")
plt.xlabel("Armour Value")
plt.title("Plot for Armour formula i.e. \nPercentage of Damage = (100 / (100 + totalArmour))")
```
